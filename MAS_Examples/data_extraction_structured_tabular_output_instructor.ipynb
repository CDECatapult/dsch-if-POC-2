# Showcasing of data extraction and creating structered tabular output
Using OpenAI or function calling open-source LLMs compatible with OpenAI. 

Structured output declarations are based on the pydantic class of Python, which enables validation of the format.

In this example, we use Wikipedia's Egyptian pyramids to extract information about the names of pyrmaids, built year and other relevant information in tabular format.


```python
import instructor
from pydantic import Field,BaseModel
from typing import List, Any
import os
from openai import OpenAI
```

```python
os.environ["OPENAI_API_KEY"] = "YOUR KEY HERE"
client = instructor.from_openai(OpenAI(), mode=instructor.Mode.JSON, )

class RowData(BaseModel):
    row: List[Any] = Field(..., description="The values for each row")


class PandasDataframe(BaseModel):
    """
    Class representing a dataframe. This class is used to convert
    data into a frame that can be used by pandas.
    """

    data: List[RowData] = Field(
        ...,
        description="Correct rows of data aligned to column names, Nones are allowed",
    )
    columns: List[str] = Field(
        ...,
        description="Column names relevant from source data, should be in snake_case",
    )

    def to_pandas(self):
        import pandas as pd

        columns = self.columns
        data = [row.row for row in self.data]

        return pd.DataFrame(data=data, columns=columns)


def dataframe_by_LLM(data: str) -> PandasDataframe:
    
    completion = client.chat.completions.create(
        model="gpt-4o",
        # model="llama3",
        # model="gorilla-openfunctions-v2-q6_K.gguf",

        temperature=0,
        response_model=PandasDataframe,
        messages=[
            {
                "role": "system",
                "content": """Map the pyramids, their build date, builders into a dataframe and correctly define the correct columns and rows""",
            },
            {
                "role": "user",
                "content": f"{data}",
            },
        ],
        max_tokens=2000,
    )    
    

    return completion


```
